## Correlation of Packet Delay and Loss in the Internet

### Método de regresión lineal 
El primer problema es **la desviación del reloj**. Si los **relojes** en el **origen y el destino no están sincronizados** y tienen un
deriva de frecuencia, las mediciones de retardo muestran un **aumento o disminución gradual** con el tiempo **dependiendo de qué
el reloj es más rápido.**

La Figura 2 muestra **la marca de tiempo** del paquete versus **el retraso por paquete**, mostrando **una disminución gradual** de retraso en el tiempo. En este caso, **el reloj del emisor es más rápido que el reloj del receptor.** 

Usamos **regresión lineal** para filtrar dicha desviación del reloj en las mediciones de retardo. **Los resultados filtrados** para los datos **sin procesar** que se muestran en **la Figura 2** son se muestra en la Figura 3.
